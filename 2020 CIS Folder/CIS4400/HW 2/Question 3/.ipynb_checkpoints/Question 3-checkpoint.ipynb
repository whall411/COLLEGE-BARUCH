```python
df = pd.read_csv('hw2q3.csv')
```

```python
df.head()
```

```python
df = df[['job__','doc__','borough','house__','street_name','job_type','job_status_descrp','landmarked','adult_estab','latest_action_date','existing_occupancy','proposed_occupancy','owner_type','job_description','initial_cost']]
```

```python
df = df.rename(columns={'job__': 'job','doc__':'doc','borough':'borough','house__':'house','street_name':'street_name','job_type':'job_type','job_status_descrp':'job_status_descrp','landmarked':'landmarked','adult_estab':'adult_estab','latest_action_date':'latest_action_date','existing_occupancy':'existing_occupancy','proposed_occupancy':'proposed_occupancy','owner_type':'owner_type','job_description':'job_description','initial_cost':'initial_cost'})
```

```python
df.head()
```

```python
df.dropna(subset=['landmarked'],inplace=True)
```

```python
df.dropna(subset=['adult_estab'],inplace=True)
```

```python
df.dropna(subset=['job_description'],inplace=True)
```

```python
df.head(100)
```

```python
# converting initial_cost to tens of thousands of dollars
```

```python
# 1st try (to_numeric)
```

```python
df['initial_cost'] = pd.to_numeric(df['initial_cost'])
```

```python
# 2nd try (downcasting)
```

```python
df['initial_cost'] = pd.to_numeric(df['initial_cost'],downcast='integer')
```

```python
# 3rd try (astype)
```

```python
df['initial_cost'] = df['initial_cost'].astype(int)
```

```python
df['initial_cost']
```

```python
# 4th try (infer_objects)
```

```python
df['initial_cost'] = df['initial_cost'].infer_objects()
```

```python
def format(x):
    return "${:.1f}K".format(x/1000)
```
