```
#scraping table from https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population
import requests
from bs4 import BeautifulSoup
```

```
url = "https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population"


html_content = requests.get(url)

soup = BeautifulSoup(html_content.content, "html.parser")

print(soup.prettify())
```

```
# use the 'find_all' function to bring back all instances of the 'table' tag in the HTML and store in 'all_tables' variable
all_tables=soup.find_all("table")
all_tables
```

```
right_table = soup.find('table', class_='wikitable sortable')

print(right_table)
```

```
#Create lists for as many columns in the table
L1 = []
L2 = []
L3 = []
L4 = []
L5 = []
L6 = []
L7 = []
L8 = []
L9 = []
L10 = []
L11 = []

for row in right_table.findAll('tr'):
  cells=row.findAll('td')
  
  if len(cells)== 11:
    L1.append(cells[0].find(text=True).rstrip("\n"))
    L2.append(cells[1].find(text=True).rstrip("\n"))   
    cells[2].span.decompose()
    L3.append(cells[2].find(text=True).rstrip("\n"))
    L4.append(cells[3].find(text=True).rstrip("\n"))
    L5.append(cells[4].find(text=True).rstrip("\n"))
    L6.append(cells[5].find(text=True).rstrip("\n"))
    L7.append(cells[6].find(text=True).rstrip("\n"))
    L8.append(cells[7].find(text=True).rstrip("\n"))
    L9.append(cells[8].find(text=True).rstrip("\n"))
    L10.append(cells[9].find(text=True).rstrip("\n"))
    L11.append(cells[10].find(text=True).rstrip("\n"))
```

```
import pandas as pd
df=pd.DataFrame()
df['2019 Rank']                     = L1
df['City']                          = L2
df['State']                         = L3
df['2019 Estimate']                 = L4
df['2010 Census']                   = L5
df['Change']                        = L6
df['Land Area sq miles']            = L7
df['Land Area sq km']               = L8
df['population desity (sq miles)']  = L9
df['population desity (sq km)']     = L10
df['location']                      = L11


df
```

```
df.to_csv("american cities by population.csv")
```
