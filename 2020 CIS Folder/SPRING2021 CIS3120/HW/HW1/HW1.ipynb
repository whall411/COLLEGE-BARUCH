```python
from bs4 import BeautifulSoup as soup
import requests
from statistics import mean
```

## Scraping Baruch Men's Swimming Team

```python
# Scraping the baruch male swimming roster url
url = "https://athletics.baruch.cuny.edu/sports/mens-swimming-and-diving/roster"
r = requests.get(url)
rtext = r.text
# Using BeautifulSoup to parse html
mswim_soup = soup(rtext,"html.parser")
print(mswim_soup.prettify())
```

```python
# Finding the swimmer heights
mswim = mswim_soup.findAll("span",{"class":"sidearm-roster-player-height"})

# loop the heights and convert to inches
def mswim_height():
    out = []
    for i in mswim:
        height_split = i.text.split( "\'")
        # convert height to inches
        feet = int(height_split[0])
        i1 = height_split[1]
        inch = i1.replace('"'," ")
        inches = int(inch)
        height = int(feet*12 + inches)
        out.append(height)
    return(out)

# Finding the length of the result
len(mswim_height())
```

 ###                                Printing average of Baruch Men's Swimming team

```python
msavg = mean(mswim_height())
print("The average height of the Baruch Men's Swimming team is","%.2f" % msavg)
```
