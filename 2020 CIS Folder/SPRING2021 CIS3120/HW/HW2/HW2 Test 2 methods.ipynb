# List of NYCHA Developments in New York City

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# finding the tables I want to scrape from
url = "https://en.wikipedia.org/wiki/List_of_New_York_City_Housing_Authority_properties"
r = requests.get(url)
rtext = r.text
# Using BeautifulSoup to parse html
nychaSoup = BeautifulSoup(rtext,"lxml")
print(nychaSoup.prettify())
```

```python
table = nychaSoup.find("table",class_="wikitable")
```

```python
for items in table.find_all("tr"):
    data = [' '.join(item.text.split()) for item in items.find_all(['th','td'])]
    print(data)
```

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/List_of_New_York_City_Housing_Authority_properties')
```

```python
for i, table in enumerate(tables):
    table.to_csv('NYCHA{}.csv'.format(i))
```

```python
df = pd.DataFrame(data = tables)
```

```python
import urllib
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/List_of_New_York_City_Housing_Authority_properties'

html_table = urllib.request.urlopen(url).read()

# fix HTML
soup = BeautifulSoup(html_table, "html.parser")
# warn! id ratings-table is your page specific
for table in soup.findChildren(attrs={'id': 'ratings-table'}): 
    for c in table.children:
        if c.name in ['tbody', 'thead']:
            c.unwrap()
```

```python
df = pd.read_html(str(soup), flavor="bs4")
len(df[0])
```
