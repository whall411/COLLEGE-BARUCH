Yahoo Finance API will need 3 mandatory arguments in this order:


- Tickers (1)
- Start date + End date or Period (2)
- Interval (3)

### example: yf.download(tickers="arg1",period="arg2",interval="arg3")

```python
# The interval is 5 min
data = yf.download(tickers='UBER', period='5d', interval='5m')
data
```
