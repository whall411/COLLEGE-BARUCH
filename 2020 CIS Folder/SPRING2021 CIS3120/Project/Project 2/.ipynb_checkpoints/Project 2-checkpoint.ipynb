```python
import numpy as np
import pandas as pd
import yfinance as yf
```

### DF1 Webscraping S&P Companies

```python
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
table = pd.read_html(url)
table = table[0]
```

```python
# Selecting first 30 rows of data
df1 = table[0:30]
```

```python
df1
```

### DF2 Using API Information

```python
# Using Yahoo Finance API to store company stock data
```

```python
# Selecting the 'Symbol' column and converting to a list
symbol = df1['Symbol'].tolist()
symbol
```

```python
df2 = yf.download(tickers=symbol, period='10d', interval='1d')
```

```python
df2 = df2['Open'].T
```

```python
open1 = "Open"
df2['Info'] = open1
```

```python
df1insert = df1[['Security','Symbol']].sort_values(by=['Symbol'],ascending=True)
```

```python
type(df1insert)
```

```python
type(df2)
```

```python
df1insert.reset_index(drop=True,inplace=True)
df2.reset_index(drop=True,inplace=True)
```

```python
df3open = pd.concat([df1insert,df2],axis=1)
```
