# Get swimmer and volleyball data for which athletes are taller

```python
import requests
from bs4 import BeautifulSoup as soup
from statistics import mean
```

### Swimming height

```python
url = "https://athletics.baruch.cuny.edu/sports/mens-swimming-and-diving/roster"
r = requests.get(url)
rtext = r.text
```

```python
print(rtext)
```

```python
swim_soup = soup(rtext,"html.parser")
swim = swim_soup.findAll("span",{"class":"sidearm-roster-player-height"})
```

```python
# loop the heights and convert to inches
def swim_height():
    out = []
    for i in swim:
        height_split = i.text.split( "\'")
        # convert height to inches
        feet = int(height_split[0])
        i1 = height_split[1]
        inch = i1.replace('"'," ")
        inches = int(inch)
        height = int(feet*12 + inches)
        out.append(height)
    return(out)
```

```python
mean(swim_height())
```
