```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
```

### IBM

```python
key = 'RY6B4DW8S5MOP3J5'
ticker = 'IBM'
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={}&apikey={}'.format(ticker, key)
```

```python
response = requests.get(url)
data = response.json()
```

```python
fm = json.dumps(data, sort_keys = True, indent = 0)
d = json.loads(fm)
```

```python
d
```

```python
ibm = pd.json_normalize(d['Time Series (Daily)'])
```

```python
ibm = ibm.T
```

```python
ibm = pd.DataFrame(ibm)
```

```python
ibm
```

```python
ibm.columns = ['IBM']
ibm
```
